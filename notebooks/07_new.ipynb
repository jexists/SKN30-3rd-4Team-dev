```python
import os
import json
import glob

def clean_and_save_to_new_dir(source_dir, output_dir):
    # 출력을 저장할 new 폴더가 없으면 생성
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        print(f"📂 새 폴더를 생성했습니다: {output_dir}")

    # 소스 폴더 내의 모든 json 파일 가져오기
    json_files = glob.glob(os.path.join(source_dir, "*.json"))
    
    if not json_files:
        print(f"[{source_dir}] 경로에 전처리할 json 파일이 존재하지 않습니다.")
        return

    print(f"총 {len(json_files)}개의 파일을 전처리하여 새 폴더로 저장합니다...")

    for file_path in json_files:
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                data = json.load(f)
            
            # 리스트 형태의 데이터인지 확인 후 필터링
            if isinstance(data, list):
                original_count = len(data)
                # "빈/표지 페이지" 텍스트가 없는 항목만 남기기
                cleaned_data = [
                    page for page in data 
                    if "빈/표지 페이지" not in page.get("page_content", "")
                ]
                
                # 새 폴더에 저장할 파일 경로 설정 (파일명은 동일하게 유지)
                file_name = os.path.basename(file_path)
                new_file_path = os.path.join(output_dir, file_name)
                
                with open(new_file_path, "w", encoding="utf-8") as f:
                    json.dump(cleaned_data, f, ensure_ascii=False, indent=4)
                
                print(f"✔ 저장 완료: {file_name} ({original_count}p -> {len(cleaned_data)}p)")
            else:
                print(f"❌ 포맷 불일치 (List 아님): {os.path.basename(file_path)}")
                
        except Exception as e:
            print(f"🔺 오류 발생 ({os.path.basename(file_path)}): {e}")

# 실행 코드
if __name__ == "__main__":
    # 소스 폴더와 새로 저장할 폴더 경로 설정
    source_directory = "../data/03_processed/one"
    output_directory = "../data/03_processed/new"
    
    clean_and_save_to_new_dir(source_directory, output_directory)
```
